<h1> UPDATE, UPGRADE, and IMPROVE SEARCH CAPABILITIES</h1>
Netflix’s search engine is very responsive and can search all titles, people, and genres, just by typing the movie titles or the actor’s name. While many users are impressing and happy with the search capabilities and features provided by Netflix, there always is a room to improve. As we can see below in the screenshot, the search for “bruce” results in anything that relates to “bruce” either from title, actor name, or genres. Netflix also provide the popular results based on the keyword typed in search by showing actor names, movie/tv show titles at the top of the page named: “Explore titles related to “. While this is helpful and can be useful, many users are not even aware of it as the users’ attention skip that line of text and go straight to the result that show the name of the title and the thumbnail as they are more attractive than that one line of text. 

    To improve the search result, instead of showing that line of text “Explore titles related to”, Netflix could use what Google’s search engine is doing, by showing the list of possible and relatable results as a dropdown option. The result should include both the titles of the movie/ tv-show as well as all the available actor names.
